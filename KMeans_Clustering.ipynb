```python
import numpy as np

np.sqrt(200)
```

```python
#Elbow Method
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
k = range(2,15)

sse = []
for i in k:
  model_demo = KMeans(n_clusters=i,random_state=0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)
plt.scatter(k,sse)
plt.plot(k,sse)
```

```python
#Silhouette score Method
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Mall_Customers.csv')
df
```

```python
x = df.iloc[:,3:5].values
x
```

```python
df.info()
```

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
```

```python

np.sqrt(200)
```

```python
from sklearn.metrics import silhouette_score
k = range(2,15)
for i in k:
  model_demo = KMeans(n_clusters=i,random_state=0)
  model_demo.fit(x)
  y_pred = model_demo.predict(x)
  print(f"{i} Clusters, Score ={silhouette_score(x,y_pred)}")
  plt.bar(i,silhouette_score(x,y_pred))

```

```python
k = 5

model = KMeans(n_clusters=k,random_state=0)
model.fit(x)

```

```python
y  =model.predict(x)
y
```

```python
y.size
```

```python
x[y == 1,1]
```

```python
np.unique(y,return_counts= True)
```

```python
#Visualization

plt.figure(figsize= (9,5))
for i in range(k):
  plt.scatter(x[y ==i,0],x[y ==i,1], label = f'Cluster {i}')
  plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s =100,c = 'skyblue',label = 'Centroids')
plt.legend()
```
